For example, there are players who shoot better contested than open, so these players would have a case for being great shooters no matter the situation. In order to account for this, I am going to make my own personal shot quality rating. 

The general rating be based off the averages above. It will be a score out of the average wide open shot percentage(times 100) - the average contested shot percentage(times 100), leading to about a 10. a perfect score would be a players that takes all open shots, while a score of 0 will be all contested. All open shots would be where that percentage falls in the middle of the 2. 

I will make one stat just applying this to all players, which will just judge and reward players for taking more open shots. However, I then want to weight this by person. Some players shoot WAYYY better when open or wide open than contested, and some shoot about the same(all over small sample sizes ofc but the general idea should persist). The farther a player strays from the average, the heavier the shot quality metric will be. For example, if you shot 0% on contested 3s and 100% on wide open 3s, this metric should penalize you a lot if you take a lot of contested threes and help you a lot if you take a lot of wide open threes. A player who shoots about the same will not see the shot quality effect their overall score much.
